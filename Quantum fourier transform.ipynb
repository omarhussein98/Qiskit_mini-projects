```python
from math import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
n=9
circuit=QuantumCircuit(n,n)
count=0
while(count<n):
    circuit.h(count)
    h=n+1-count
    k=count+1
    count1=2
    while(count1<h):
        circuit.crz(2*pi/(2**(count1)),k,count)
        count1=count1+1
        k=k+1
    count=count+1
    circuit.barrier()
circuit.draw(output='mpl')
```
