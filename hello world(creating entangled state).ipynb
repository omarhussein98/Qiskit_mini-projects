```python
from qiskit import *
```

```python
%matplotlib inline
```

```python
qubit=QuantumRegister(2)
```

```python
cbit=ClassicalRegister(2)
```

```python
circuit=QuantumCircuit(qubit,cbit)
```

```python
circuit.draw()
```

```python
circuit.h(qubit[0])
```

```python
circuit.draw(output='mpl')
```

```python
circuit.cx(qubit[0],qubit[1])
```

```python
circuit.draw(output='mpl')
```

```python
circuit.measure(qubit[0],cbit[1])
```

```python
circuit.draw(output='mpl')
```

```python
circuit.measure(qubit[1],cbit[0])
```

```python
circuit.draw(output='mpl')
```

```python
simulator=Aer.get_backend('qasm_simulator')
```

```python
result=execute(circuit, simulator).result()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(result.get_counts(circuit))
```

```python
IBMQ.load_account()
```

```python
job=execute(circuit, qcomp)
```
