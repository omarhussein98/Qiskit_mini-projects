```python
from qiskit import *
```

```python
from math import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
n=4
circuit=QuantumCircuit(n,n)
count=0
while(count<n):
    circuit.h(count)
    h=n+1-count
    k=count+1
    count1=2
    while(count1<h):
        circuit.crz(2*pi/(2**(count1)),k,count)
        count1=count1+1
        k=k+1
    count=count+1
    circuit.barrier()
measurement=0
while(measurement<n):
    circuit.measure(measurement,measurement)
    measurement=measurement+1
simulator=Aer.get_backend('qasm_simulator')
result=execute(circuit,simulator).result()
plot_histogram(result.get_counts())
#approximately equal probabilities since qft just changes the phase
```

```python
circuit.draw(output='mpl')
```
